# outlier IQR Method

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(data=df)
plt.xticks(rotation=90)
```

# dengan menggunakan boxplot yang basisnya adalah IQR dapat kita lihat bahwa kolom score dan GDP tidak memiliki nilai yang outlier,mari kita cari tau untuk kolom lain berapa banyak outliernya dan apakah itu outlier yang bagus atau tidak

```python
IQR = df['Score'].describe()['75%']-df['Score'].describe()['25%']
p25 = df['Score'].describe()['25%']
p75 = df['Score'].describe()['75%']
lwOl = p25 - 1.5*IQR
upOl = p75 + 1.5*IQR
dfol = df[(df['Score']>upOl) | (df['Score']<lwOl)]
dfol
```

```python
IQR = df['GDP per capita'].describe()['75%']-df['GDP per capita'].describe()['25%']
p25 = df['GDP per capita'].describe()['25%']
p75 = df['GDP per capita'].describe()['75%']
lwOl = p25 - 1.5*IQR
upOl = p75 + 1.5*IQR
dfol = df[(df['GDP per capita']>upOl) | (df['GDP per capita']<lwOl)]
dfol
```

```python
IQR = df['Social support'].describe()['75%']-df['Social support'].describe()['25%']
p25 = df['Social support'].describe()['25%']
p75 = df['Social support'].describe()['75%']
lwOl = p25 - 1.5*IQR
upOl = p75 + 1.5*IQR
dfol = df[(df['Social support']>upOl) | (df['Social support']<lwOl)]
dfol
```

# untuk kolom lainnya saya menggunakan distribution plot untuk mengetahui apakah datanya terdistribusi normal atau tidak,atau skew,atau tidak dengan menggunakan skew test,mari kita lihat

```python
from scipy.stats import skew, kurtosis
```

```python
# kalo skew lebih dari satu berati ekornya panjang banget kalo positif kekanan
# yg simetris tuh,-0,5 sampai 0,5
# If the skewness is between -0.5 and 0.5, the data are fairly symmetrical
# If the skewness is between -1 and – 0.5 or between 0.5 and 1, the data are moderately skewed
# If the skewness is less than -1 or greater than 1, the data are highly skewed
```

```python
# kalo kurtosis lebih dari 0 kemungkinan tailnya besar,sangat gemuk,leptokurtic
# kalo kurang dari 0 platiurtic agak lebih kurus,kalo di pdf lebih tinggi mendekatai meannya
# kalo mendekatai 0,mesokurtic
# If the kurtosis is close to 0, then a normal distribution is often assumed.  These are called mesokurtic distributions.  
# If the kurtosis is less than 0, then the distribution is light tails and is called a platykurtic distribution.  
# If the kurtosis is greater than 0, then the distribution has heavier tails and is called a leptokurtic distribution.
```
